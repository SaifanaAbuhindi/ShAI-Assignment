# Data Cleaning

**Handling missing data: we know that Columns With Missing Values are 'BasePay', 'OvertimePay', 'OtherPay', 'Benefits', 'Notes', and 'Status' based on the work done above, missing values will be handled apropriately based on each column and it's values.**

```python
# handling missing values in 'BasePay' column
df.loc[df['BasePay'].isnull()].head(5)
```

```python
df.loc[df['BasePay'].isnull()].tail(5)
```

### we can notice there are rows that barely have any information and since there are no job titles, no names, and no total pay so we will be dropping them

```python
#counting the rows that have 'Not Provided' as 'JobTitle'
len(df[df['JobTitle'] == 'Not provided'])
```

```python
#dropping the rows with unprovided names and job titles
df.drop(df[df['JobTitle'] == 'Not provided'].index, inplace = True)
```

## now the rest of the null values in the column 'BasePay' will be filled with the values from the column 'TotalPay' minus the values from 'OtherPay' and 'OvertimePay'

```python
#filling the null values with the difference between 'TotalPay' and the sum of 'OtherPay' and 'OvertimePay'
df.loc[df['BasePay'].isnull(), 'BasePay'] = df.loc[df['BasePay'].isnull(), 'TotalPay'] - (df.loc[df['BasePay'].isnull(), 'OtherPay'] + df.loc[df['BasePay'].isnull(), 'OvertimePay'])
```

```python
df['BasePay'].isnull().sum()
```

```python
df.head()
```

```python
#checking the rest of the columns again for null values

#used this option to display all the columns null values counts
pd.set_option('display.max_rows', None)
#displaying the columns null values counts
df.isnull().sum()
```

```python
#resetting the max display to 10
pd.set_option('display.max_rows', 10)
```

```python
# handeling null values in 'Benefits' column 
df.loc[df['Benefits'].isnull()]
```

## since there is two columns 'TotalPay' and 'TotalPayBenefits' and they have 0 null values so they are suitable to be used in order to fill in the the null values in the  'Benefits' column 

```python
# filling the nulls in 'Benefits' column with the difference between TotalPayBenefits and TotalPay for the rows with missing values
df.loc[df['Benefits'].isnull(), 'Benefits'] = df.loc[df['Benefits'].isnull(), 'TotalPayBenefits'] - df.loc[df['Benefits'].isnull(), 'TotalPay']
```

```python
#checking the rest of the columns again for null values

#used this option to display all the columns null values counts
pd.set_option('display.max_rows', None)
#displaying the columns null values counts
df.isnull().sum()
```

```python
df.info()
```

## in the above info it is shown that the columns 'Notes' and 'Status' are totally empty columns and so they will be dropped

```python
#dropping the columns 'Status' and 'Notes'
df.drop(['Status', 'Notes'], axis=1, inplace=True)
```

# Basic Data Visualization

### Visualizing the distribution of salaries (TotalPay).

```python
import matplotlib.pyplot as plt

# Histogram for distribution of salaries
df['TotalPay'].hist()

# adding title and labels for the graph
plt.ylabel('Salary')
plt.xlabel('Number of Employees With The Salary')
plt.title('Distribution of Salaries')
```

**as noticed in the histogram the most occuring salaries are between 35000 and 50000**

### Representing the proportion of employees in different departments
**since there is no _Department_ column the below pie chart will visualize the top 7 job titles and how much do they represent of the total employees**

```python
#pie plot for percentage of employees in departments

Jobs = df['JobTitle'].value_counts()

# Selecting the top 7 Jobs/departments
top_Jobs = Jobs.head(7)

# Calculating the count for 'Others' by summing the counts of remaining Jobs
top_Jobs['Others'] = Jobs.sum() - top_Jobs.sum()

#plotting the pie chart
plt.pie(top_Jobs, autopct='%.2f%%', 
        pctdistance=1.16, # Moving percentages outside the circle
        wedgeprops=dict(width=0.3, edgecolor='w'))

#making a legend and specifying its' position 
plt.legend( labels = top_Jobs.index, loc='upper left', 
           bbox_to_anchor=(1.2, 1))
#title 
plt.title('Proportions of Employees in Different Departments', fontsize=15)
```
